NumPy est une bibliothèque essentielle pour le calcul scientifique en Python. Elle fournit un support pour les tableaux multidimensionnels (appelés ndarray), ainsi que des fonctions mathématiques sophistiquées pour les manipuler.

Voici quelques concepts clés et exemples d'utilisation de NumPy :

```python
# Créer un tableau à partir d'une liste Python
a = np.array([1, 2, 3, 4, 5])
b = [1,4,5]
b2 = np.array(b)
print(a)
print(b)
```

```python
# Tableau de zéros
zeros = np.zeros((3, 3))
print(zeros)

# Tableau de uns
ones = np.ones((2, 4))
print(ones)

# Tableau d'une séquence de nombres
arange = np.arange(0, 10, 2)
print(arange)
```

```python
c = np.array([10, 20, 30, 40, 50])

# Addition
sum_result = a + c
print(sum_result)

# Multiplication
mul_result = a * c
print(mul_result)
```

```python
# Accès à un élément
element = a[2]
print(element)

# Tranche (slicing)
slice_a = a[1:4]
print(slice_a)
```

```python
# Moyenne
mean_a = np.mean(a)
print(mean_a)

# Somme
sum_a = np.sum(a)
print(sum_a)
```
